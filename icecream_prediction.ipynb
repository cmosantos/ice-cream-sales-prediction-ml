```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = {
    "Temperature":[18,20,22,24,26,28,30,32,34,36],
    "Sales":[80,100,120,140,160,180,200,220,240,260]
}

df = pd.DataFrame(data)

X = df[["Temperature"]]
y = df["Sales"]

model = LinearRegression()
model.fit(X, y)

prediction = model.predict(pd.DataFrame([[29]], columns=["Temperature"]))

print("Prediction for 29°C:", prediction[0])

plt.scatter(df["Temperature"], df["Sales"])
plt.xlabel("Temperature")
plt.ylabel("Ice Cream Sales")
plt.show()
```
